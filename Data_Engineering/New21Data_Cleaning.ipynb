## Can the fullname of guarantor be a matching key?

```python
# check for basic information of id-type columns
def validate_name_column(data, column_name):
    # Check for missing values
    missing_values = data[column_name].isnull().sum()
    
    # Check for empty strings or invalid entries (non-numeric values, if applicable)
    invalid_values = data[column_name].apply(lambda x: isinstance(x, str) and x.strip() == "").sum()
    
    # Check for duplicate IDs
    duplicate_ids = data[column_name].duplicated().sum()
    
    # Print the results
    print(f"# of missing name: {missing_values}")
    print(f"Invalid values: {invalid_values}")
    print(f"Duplicate values: {duplicate_ids}")
```

```python
validate_name_column(patient, 'guarantor')
```

```python
validate_name_column(guarantor, 'firstname')
```

```python
validate_name_column(guarantor, 'lastname')
```

## Try to use guarantor name as a matching key

```python
# Create a copy of the guarantor dataset
data_guarantor = guarantor.copy()

data_guarantor['guarantor_fullname'] = (
    data_guarantor['firstname'].fillna('') + ' ' +
    data_guarantor['middlename'].fillna('') + ' ' +
    data_guarantor['lastname'].fillna('')
)

# Remove any extra spaces if 'middlename' is missing
data_guarantor['guarantor_fullname'] = data_guarantor['guarantor_fullname'].str.strip()

data_guarantor[['firstname', 'middlename', 'lastname', 'guarantor_fullname']].head()

```

```python
# Check if the name could be matched

# Standardize the text in both datasets (convert to lowercase and strip spaces)
patient['guarantor_clean'] = patient['guarantor'].str.lower().str.strip()
data_guarantor['guarantor_fullname_clean'] = data_guarantor['guarantor_fullname'].str.lower().str.strip()

# Perform the match
matches = patient['guarantor_clean'].isin(data_guarantor['guarantor_fullname_clean'])

# Add a column to the patient dataset indicating match status
patient['Match_Status'] = matches

# Check the results
print(f"Total rows in patient dataset: {len(patient)}")
print(f"Number of matches found: {matches.sum()}")
print(f"Number of unmatched rows: {len(patient) - matches.sum()}")
```

```python
# Merge the data if the name could match (for test)
data_patient = patient.copy()
merged_data = pd.merge(data_patient, data_guarantor, 
    left_on='guarantor_clean', 
    right_on='guarantor_fullname_clean', 
    how='inner',  # Use 'inner' to keep only matched rows
    suffixes=('_patient', '_guarantor')
)


print("Merged dataset:")
merged_data.head()
```

## Validation check through Patient data Birth_date 

```python
from datetime import datetime
today = datetime.today()

patient['birth_date'] = pd.to_datetime(patient['birth_date'], errors='coerce')
patient['age'] = patient['birth_date'].apply(lambda x: today.year - x.year - ((today.month, today.day) < (x.month, x.day)) if pd.notnull(x) else None)
filtered_patients = patient[(patient['age'] <= 3) | (patient['age'] >= 80)]

print("Filtered dataset with valid ages (3 to 80):")
print(filtered_patients[['id', 'fullname', 'birth_date', 'age']])
```
